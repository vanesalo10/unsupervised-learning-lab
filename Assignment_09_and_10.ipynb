### UMAP

UMAP (Uniform Manifold Approximation and Projection) is a dimensionality reduction technique that maps high-dimensional data to a lower-dimensional space.

At a high level, UMAP constructs a graph representation of the data, where each data point is a node and edges connect nodes that are close in high-dimensional space. It then optimizes a low-dimensional embedding of the data by minimizing a cost function that balances preserving local neighborhood structure in the high-dimensional space and ensuring that the embedding is uniformly distributed across the low-dimensional space.

### LDA

Linear discriminant analysis (LDA) is a classical statistical technique used for supervised dimensionality reduction and classification. It is based on a mathematical framework that combines ideas from linear algebra, probability theory, and optimization.

At a high level, LDA finds a linear transformation of the data that maximizes the separation between classes while preserving the information in the original data. Specifically, it finds a projection that maximizes the ratio of between-class variance to within-class variance.

The underlying mathematics of LDA involves modeling the distribution of the data using multivariate normal distributions, and then finding the projection that maximizes the Fisher criterion, a measure of the separation between classes. This optimization can be performed using eigenvalue decomposition or singular value decomposition.

LDA is useful for a variety of applications in machine learning and data analysis, including dimensionality reduction, feature extraction, and classification. By reducing the dimensionality of the data while preserving the information relevant for classification, LDA can help to improve the performance of machine learning models and reduce computational costs. It is commonly used in image and signal processing, as well as in fields such as finance and biomedical engineering.
